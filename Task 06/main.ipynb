```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df=pd.read_csv("House Price Prediction Dataset.csv")
df=df.dropna(axis=0)
m={'Yes':1,'No':0}
df['Garage']=df['Garage'].map(m)
m={'Excellent':3,'Good':2,'Fair':1,'Poor':0}
df['Condition']=df['Condition'].map(m)
for j in df['Location'].unique():
    df[f"Location-{j}"]=(df["Location"]==j).astype(int)
df=df.drop(columns=['Location'])
cols=['Area',"Bedrooms","Floors","YearBuilt","Condition","Price"]
scaler=StandardScaler()
df[cols]=scaler.fit_transform(df[cols])
df.head(5)
```

```python
x=df.drop(columns=['Id','Price'])
y=df['Price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
predict=model.predict(x_test)
print(f"SCORE: {r2_score(y_test,predict)}")
print(f"MEAN ABSOLUTE ERROR: {mean_absolute_error(y_test,predict)}")
print(f"ROOT MEAN SQUARED ERROR: {root_mean_squared_error(y_test,predict)}")
```

```python
model=GradientBoostingRegressor()
model.fit(x_train,y_train)
predict=model.predict(x_test)
print(f"SCORE: {r2_score(y_test,predict)}")
print(f"MEAN ABSOLUTE ERROR: {mean_absolute_error(y_test,predict)}")
print(f"ROOT MEAN SQUARED ERROR: {root_mean_squared_error(y_test,predict)}")
```
