```python
from huggingface_hub import InferenceClient

Tok="Your token over here"
client = InferenceClient(api_key=Tok)
limitations = """
You are a friendly, comforting and clear General Health Query Chatbot and you are required to answer questions related to that. 
Strict Rules to follow:
1. You will politely refuse to answer any irrelevant question except greeting messages
2. Always include a disclaimer advising the user to not rely on your answers finally, and consult a doctor for final decision. 
3. Do NOT advise any specific medication dosage and diagnosis. 
4. Keep answers to the point, easier to read in simple english and clear. 
5. When the user input 'quit' you will give a comforting goodbye response
6. Do not give harmful responses
"""
print("\nHEALTH CHAT BOT LIVE (enter 'quit' to exit)\n")
while True:
    usr=input()
    print(f"YOU: {usr}\n")
    res=client.chat.completions.create(
    model="Qwen/Qwen2.5-72B-Instruct:novita",
    messages=[
        {"role":"user","content":usr},
        {"role":"system","content":limitations} 
    ])
    print(f"BOT: {res.choices[0].message.content}")
    print("=========================================================================================")
    if usr.lower()=="quit":
        break
```
